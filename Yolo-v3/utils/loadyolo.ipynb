```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
pd.set_option("display.width", 400)
```

```python
path = Path("../data/train_val.npy")
arr = np.load(path, allow_pickle=True)
df = pd.DataFrame(arr)
df.columns = ["filename", "class", "box"]
df.head()
```

```python
df = df.explode("class").explode("box", ignore_index=True)
df.head()
```

```python
box = pd.DataFrame(df["box"].tolist())
box.columns = ["xmin", "ymin", "xmax", "ymax"]
box.head()
```

```python
df = pd.concat([df.loc[:, "filename":"class"], box], axis=1)
df.head()
```
