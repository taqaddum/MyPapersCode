```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
pd.set_option("display.width", 400)
```

```python
path = Path("../data/train_val.npy")
arr = np.load(path, allow_pickle=True)
a = np.array(arr[0][1], dtype=np.dtype('S40, f4, f4, f4, f4'))
np.random.shuffle(a)
```

```python
df = pd.DataFrame(arr)
df.columns = ["filepath", "boxclass"]
df.head()
```

```python
df = df.explode("boxclass", ignore_index=True)
df.head()
```

```python
box = pd.DataFrame(df["boxclass"].tolist())
box.columns = ["class", "xmin", "ymin", "xmax", "ymax"]
box.head()
```

```python
df = pd.concat([df.loc[:, "filepath":"filename"], box], axis=1)
df.head()
```
